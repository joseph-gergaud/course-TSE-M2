# Descriptive Statistics
## Descriptive Statistics in one dimension for Continuous variables
### Histograms and Kernel Density Estimation

```julia
using Plots, StatsPlots

using RDatasets
iris = RDatasets.dataset("datasets", "iris")  # Iris Datas
println("propertynames(iris) = ", propertynames(iris))
p1 = histogram(iris[!,:SepalLength], norm=true, label="histogram")   # Histogram
density!(p1,iris[!,:SepalLength], c=:red, label="Density via StatsPlots") # Kernel Density Estimation
hvals = [0.5,0.2,0.1]
colors = [:green, :orange, :black]
for i in 1:length(hvals)
    density!(p1,iris[!,:SepalLength], c=colors[i], bandwidth=hvals[i],label="h=$(hvals[i])")
end
#salaries = [2500, 5000, 10000, 15000, 20000, 30000, 50000, 70000, 100000, 200000, 400000]
#relative_frequences = [0.0067, 0.3018, 0.2750, 0.1709, 0.1445, 0.0701, 0.0166, 0.0081, 0.0051, 0.0010]

plot(p1, title ="Sepal Length of Iris Datas")
```

### Box Plots and Violin Plot

```julia
p2 = boxplot(iris[!,:Species], iris[!,:SepalLength], label = "Sepal Length")
p3 = violin(iris[!,:Species], iris[!,:SepalLength], label = "Sepal Length")
plot(p2,p3)
```

## Two dimensions
This comes from the book "Statistics with Julia, Fundamentals for Data Science, Machine Learning and Artificial Intelligence", Yoni NAzarathy and Hayden Klok, Springer, 2021, pages 168-169.

```julia
using Measures

insertSpace(name) = begin
    i = findlast(isuppercase,name)
    name[1:i-1]*" "*name[i:end]
end
featureNames = insertSpace.(string.(names(iris)))[1:4]

scatters = [ 
scatter(iris[:,i], iris[:,j], c=[:blue :red :green], group=iris.Species, xlabel=featureNames[i], ylabel=featureNames[j], legend = i==1 && j==1)    
for i in 1:4, j in 1:4 ]
plot(scatters..., size=(1200,800), margin = 4mm) # ... for expanding
```
